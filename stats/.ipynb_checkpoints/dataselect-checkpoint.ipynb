```python
import MySQLdb
import time
t1 = time.clock()
conn = MySQLdb.connect(host='localhost',user='root',passwd='9182736450',db='iwe')
cur = conn.cursor()
sql = '''
SELECT COUNT(*) AS bugnum, transition,AVG(UNIX_TIMESTAMP(ts0)),AVG(UNIX_TIMESTAMP(ts1)),
AVG(UNIX_TIMESTAMP(ts2)),AVG(UNIX_TIMESTAMP(ts3)),AVG(UNIX_TIMESTAMP(ts4)),AVG(UNIX_TIMESTAMP(ts5)) FROM iwe_statustran 
WHERE ts0 > '2000-1-1' AND ts0 < '2001-1-1' 
GROUP BY (transition)
ORDER BY (bugnum) DESC
'''
cur.execute(sql)
data = cur.fetchall()
t2 = time.clock()
print t2 - t1
```
